```python
import pandas as pd

# --- KONFIGURASI NAMA FILE & KOLOM ---
file_a = 'view_trace_expired_mtv_update1.csv'
file_b = 'drive/MyDrive/Era - Loyalty/view_trace_expired_mtv_update1_lm2th_1_20260401_170654.csv'
output_file = 'missing_in_view_trace_expired_mtv_update1_FIXED.csv'

col_id_a = 'debitloyaltyledgerid'
col_id_b = 'debitloyaltyledgerid__c'

print("Sedang memproses... Harap tunggu.")

try:
    # 1. Baca File B secara UTUH (Hanya kolom ID)
    df_b = pd.read_csv(file_b, usecols=[col_id_b], dtype={col_id_b: str})
    # Simpan ke SET untuk pencarian kilat (O(1) complexity)
    ids_in_b = set(df_b[col_id_b].str.strip().dropna().unique())
    print(f"File B terbaca. Total {len(ids_in_b)} ID referensi.")

    # 2. Baca File A secara UTUH
    # low_memory=False memastikan pandas tidak menebak-nebak tipe data
    df_a = pd.read_csv(file_a, dtype={col_id_a: str}, low_memory=False)
    df_a[col_id_a] = df_a[col_id_a].str.strip() # Bersihkan spasi
    print(f"File A terbaca. Total {len(df_a)} baris untuk dicek.")

    # 3. Bandingkan: Ambil baris di A yang ID-nya TIDAK ADA di dalam set B
    data_hilang = df_a[~df_a[col_id_a].isin(ids_in_b)]

    # 4. Simpan hasil
    data_hilang.to_csv(output_file, index=False)

    print("-" * 30)
    print(f"PROSES SELESAI!")
    print(f"Total baris yang hilang (ada di A tapi tidak di B): {len(data_hilang)}")
    print(f"Hasil disimpan di: {output_file}")

except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```
